```python
import matplotlib.pyplot as plt
import data_extractor as de
```

```python
# Get data

sequences = de.get_seq_dict("rcsb_pdb_protprot_seq")
affinities = de.format_y_data("prot-prot_complexes.xls")

# print(sequences)
```

```python
# Plot distribution of number of proteins in each protein complex

fig, ax = plt.subplots()
ax.hist([len(prot) for prot in sequences.values()])
ax.set_xlabel('# Chains in Complex')
ax.set_ylabel('Frequency')
```

```python
# Plot distribution of number of amino acids in each protein complex

fig, ax = plt.subplots()
ax.hist([sum([len(seq) for seq in prot]) for prot in sequences.values()])
ax.set_xlabel('Total Chain Length (aa)')
ax.set_ylabel('Frequency')
```

```python
# Plot distribution of binding affinities

fig, ax = plt.subplots()
plt.hist(affinities)
ax.set_xlabel('$pK_d$')
ax.set_ylabel('Frequency')
```
